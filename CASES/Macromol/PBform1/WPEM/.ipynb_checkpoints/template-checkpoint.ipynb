```python
if __name__ == "__main__":

    import WPEMAlgorithm.WPEMModel as WPEMModel


    # The wavelength is set according to the actual light source
    # wavelength = [1.54059, 1.54441]
    wavelength = [1.03]
    # Enter the file name to read on the right side of the equal sign. Please remember to quote, string.
    no_bac_intensity_file = "no_bac_intensity.csv"  # The file name of non-background data (2theta-intensity data)
    original_file = "intensity.csv"  # The file name of raw/original data (2theta-intensity data)
    bacground_file = "bac.csv"  # The file name of background data (2theta-intensity data)


    # Input the initial lattice constants {a, b, c, α, β, γ}, whose values need to be assumed in advance.
    # The example given is the assumed lattice constants of crystal Tb2BaCoO5.
    Lattice_constants = [[17.53,17.53,6.47,90,90,120]]


    # Determine the latent parameters
    WPEMModel.Built().fit(wavelength, Var, Lattice_constants,no_bac_intensity_file, original_file,bacground_file,low_bound=6,
                            up_bound=16,bta=0.78, subset_number=3,LogPrint = True,iter_max = 300, asy_C = 0,Macromolecule = True,cpu = 4)
```
