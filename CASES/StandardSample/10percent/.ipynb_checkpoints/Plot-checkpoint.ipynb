```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset

# ======================
# Load data
# ======================
exp = pd.read_csv('intensity.csv', header=None)
fit = pd.read_csv('./DecomposedComponents/fitting_profile.csv', header=None)

phase_1 = pd.read_csv('./DecomposedComponents/NaCl.csv', header=None)
phase_2 = pd.read_csv('./DecomposedComponents/Li2CO3.csv', header=None)

# Data
x_exp, y_exp = exp.iloc[:, 0].values, exp.iloc[:, 1].values
x_fit, y_fit = fit.iloc[:, 0].values, fit.iloc[:, 1].values

x_p1, y_p1 = phase_1.iloc[:, 0].values, phase_1.iloc[:, 1].values
x_p2, y_p2 = phase_2.iloc[:, 0].values, phase_2.iloc[:, 1].values

diff = y_exp - y_fit

# ======================
# Range control
# ======================
theta_max = 80

def cut(x, y):
    m = x <= theta_max
    return x[m], y[m]

x_exp, y_exp = cut(x_exp, y_exp)
x_fit, y_fit = cut(x_fit, y_fit)
x_p1, y_p1 = cut(x_p1, y_p1)
x_p2, y_p2 = cut(x_p2, y_p2)
diff = diff[:len(x_exp)]

# ======================
# Style (Nature)
# ======================
plt.rcParams.update({
    "font.size": 13,
    "axes.linewidth": 1.4,
    "font.family": "sans-serif",
})

# ======================
# Layout
# ======================
fig = plt.figure(figsize=(6.5, 6))
gs = GridSpec(3, 1, height_ratios=[3.4, 2.6, 1], hspace=0.05)

ax_main = fig.add_subplot(gs[0])
ax_phase = fig.add_subplot(gs[1], sharex=ax_main)
ax_diff = fig.add_subplot(gs[2], sharex=ax_main)

# ======================
# (a) Experiment + Fit
# ======================
ax_main.plot(x_exp, y_exp, color='black', lw=1.8, label='     ')
ax_main.plot(x_fit, y_fit, color='#7A1E1E', lw=1.6, label='     ')

ax_main.legend(frameon=False, loc='upper left')

ax_main.tick_params(axis='x', bottom=False, labelbottom=False)

# ======================
# (b) Phase-resolved stacked profiles
# ======================
offset = 0.6 * max(y_p2)

# Phase 2 (bottom)
ax_phase.fill_between(
    x_p2, y_p2,
    color='#DD8452', alpha=0.55, linewidth=0
)
ax_phase.plot(
    x_p2, y_p2,
    color='#B65C2E', lw=1.3,
    label='Li$_2$CO$_3$'
)

# Phase 1 (top)
ax_phase.fill_between(
    x_p1, y_p1 + offset, offset,
    color='#4C72B0', alpha=0.55, linewidth=0
)
ax_phase.plot(
    x_p1, y_p1 + offset,
    color='#355C9A', lw=1.3,
    label='NaCl'
)

ax_phase.legend(frameon=False, loc='upper right')

ax_phase.set_yticks([])
ax_phase.tick_params(axis='x', bottom=False, labelbottom=False)

# ======================
# Inset (on MAIN panel, top-right)
# ======================
axins = inset_axes(
    ax_main,
    width="32%",
    height="42%",
    loc='upper right',
    borderpad=1.2
)

zoom_min, zoom_max = 30, 36

m1 = (x_p1 >= zoom_min) & (x_p1 <= zoom_max)
m2 = (x_p2 >= zoom_min) & (x_p2 <= zoom_max)

axins.fill_between(x_p2[m2], y_p2[m2], color='#DD8452', alpha=0.6)
axins.plot(x_p2[m2], y_p2[m2], color='#B65C2E', lw=1.1)

axins.fill_between(
    x_p1[m1], y_p1[m1] + offset, offset,
    color='#4C72B0', alpha=0.6
)
axins.plot(
    x_p1[m1], y_p1[m1] + offset,
    color='#355C9A', lw=1.1
)

axins.set_xlim(zoom_min, zoom_max)
axins.set_yticks([])
axins.tick_params(labelsize=10)
axins.set_title('     ', fontsize=10, pad=2)

mark_inset(ax_main, axins, loc1=2, loc2=4, fc="none", ec="0.5")

# ======================
# (c) Difference
# ======================
d_offset = -0.35 * np.max(np.abs(diff))
ax_diff.plot(x_exp, diff + d_offset, color='dimgray', lw=1.2)
ax_diff.axhline(d_offset, color='gray', lw=0.8)

ax_diff.set_yticks([])

ax_diff.set_xlim(x_exp.min(), theta_max)

# ======================
# Clean spines
# ======================
for ax in [ax_main, ax_phase, ax_diff]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# ======================
# Save
# ======================
plt.savefig(
    '10.png',
    dpi=600,
    bbox_inches='tight'
)
plt.show()
```
