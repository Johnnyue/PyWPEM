# 批量计算

```python
import os
import shutil
import pandas as pd
import numpy as np

file_name_set = []
lattice_set = []

Lattice_constants = [[2.87549, 2.87549, 14.18056, 90.0, 90.0, 120.0]]

for i in range(157): 
#for i in range(2): 
    index = i + 1
    file_name = f'./data/{index}.xlsx'
    

    try:
        # 复制并重命名为 intensity.xlsx（覆盖）
        shutil.copyfile(file_name, 'intensity.xlsx')

        # 读取 Excel 文件并计算
        intensity_excel = pd.read_excel(file_name, header=None)
        var = WPEM.BackgroundFit(intensity_excel, lowAngleRange=17, poly_n=8, bac_split=10, bac_num=80)

        wavelength = [1.540593, 1.544414]
        no_bac_intensity_file = "no_bac_intensity.csv"
        original_file = "intensity.xlsx"
        bacground_file = "bac.csv"


        _, _lattice = WPEM.XRDfit(
            wavelength, var, Lattice_constants, no_bac_intensity_file, original_file, bacground_file,Ave_Waves=True,
            subset_number=5, low_bound=15, up_bound=40, bta=0.85, iter_max=20, asy_C=0, InitializationEpoch=0,
        )
        lattice = _lattice[0]
        Lattice_constants = [lattice.tolist()] 
        

    except Exception as e:
        print(f'Error at index {index}: {e}')
        lattice = np.nan

    file_name_set.append(file_name)
    lattice_set.append(lattice)


# 保存结果为 Excel 表格
import pandas as pd

# 转换 lattice_set 为字符串格式，便于保存成表格（兼容 NaN 和数组）
lattice_strs = [str(l) if isinstance(l, (list, np.ndarray)) else np.nan for l in lattice_set]

df = pd.DataFrame({
    'File Name': file_name_set,
    'Lattice': lattice_strs
})

df.to_excel('lattice_results.xlsx', index=False)
print("结果已保存为 lattice_results.xlsx")
```
