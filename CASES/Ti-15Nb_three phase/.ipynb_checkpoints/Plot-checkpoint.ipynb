```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset

# ======================
# Load data
# ======================
exp = pd.read_csv('intensity.csv', header=None)
fit = pd.read_csv('./DecomposedComponents/fitting_profile.csv', header=None)
peak = pd.read_csv('./DecomposedComponents/sub_peaks.csv')['mu_i']

x_exp, y_exp = exp.iloc[:, 0], exp.iloc[:, 1]
x_fit, y_fit = fit.iloc[:, 0], fit.iloc[:, 1]
diff = y_exp - y_fit

# Only odd-indexed peaks
peak_sel = peak.iloc[::2]

# ======================
# Global style
# ======================
plt.rcParams.update({
    "font.family": "Arial",
    "font.size": 12,
    "axes.linewidth": 2.0,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "figure.facecolor": "white"
})

COLOR_EXP = '#1A1A1A'
COLOR_FIT = '#C23B2A'
COLOR_DIFF = '#6A6A6A'
COLOR_FILL = '#E6B27E'
COLOR_PEAK = '#7A7A7A'

# ======================
# Layout
# ======================
fig = plt.figure(figsize=(7.6, 3.9))
fig.subplots_adjust(left=0.10, right=0.98, top=0.96, bottom=0.14)

gs = GridSpec(2, 1, height_ratios=[8.2, 1.2], hspace=0.05)

ax = fig.add_subplot(gs[0])
ax_peak = fig.add_subplot(gs[1], sharex=ax)

# ======================
# Main pattern
# ======================
ax.plot(x_exp, y_exp, lw=0.6, alpha=0.35, color=COLOR_EXP)
ax.plot(x_exp, y_exp, lw=1.6, color=COLOR_EXP, zorder=3)
ax.plot(x_fit, y_fit, lw=1.3, color=COLOR_FIT, zorder=4)

#ax.set_ylabel('Intensity (a.u.)', fontsize=12)
ax.tick_params(labelbottom=False)

# ======================
# Difference (shifted down)
# ======================
diff_offset = y_exp.min() - 0.15 * (y_exp.max() - y_exp.min())

ax.plot(x_exp, diff + diff_offset, lw=0.9, color=COLOR_DIFF)
ax.axhline(diff_offset, lw=0.6, ls='--', color='0.6')

# ======================
# Bragg peaks (odd only)
# ======================
ax_peak.scatter(
    peak_sel,
    np.full(len(peak_sel), 0.5),
    marker='|',
    s=40,
    color=COLOR_PEAK
)

ax_peak.set_ylim(0, 1)
ax_peak.set_yticks([])
#ax_peak.set_ylabel('Bragg', fontsize=10)
#ax_peak.set_xlabel(r'$2\theta$ (degree)', fontsize=12)

# ======================
# Inset: under-peak area filling
# ======================
zoom_min, zoom_max = 20, 45

axins = inset_axes(
    ax,
    width="50%",
    height="45%",
    loc='upper right',
    borderpad=1.6
)

axins.plot(x_exp, y_exp, lw=1.5, color=COLOR_EXP)
axins.plot(x_fit, y_fit, lw=1.5, color=COLOR_FIT)

mask = (x_exp > zoom_min) & (x_exp < zoom_max)

# Local baseline (robust)
baseline = np.percentile(y_fit[mask], 5)

# Fill area under peaks
axins.fill_between(
    x_exp[mask],
    baseline,
    y_fit[mask],
    color=COLOR_FILL,
    alpha=0.5,
    linewidth=0
)

axins.set_xlim(zoom_min, zoom_max)
axins.set_ylim(
    baseline * 0.98,
    y_exp[mask].max() * 1.03
)

axins.tick_params(labelsize=8)
for s in axins.spines.values():
    s.set_linewidth(0.9)

mark_inset(ax, axins, loc1=2, loc2=4, ec='0.45', lw=0.9)

axins.tick_params(labelsize=12)

for s in axins.spines.values():
    s.set_linewidth(1.5)


# ======================
# Clean spines
# ======================
for a in (ax, ax_peak):
    a.spines['top'].set_visible(False)
    a.spines['right'].set_visible(False)


plt.savefig(
    'Ti-3phase.png',
    dpi=600,
    bbox_inches='tight'
)

plt.show()
```
